```python
print(soup.prettify())  # Print the parsed HTML
```

```python
# Find all elements and collect their classes
classes = set()
for tag in soup.find_all(class_=True):  # Select elements with a class attribute
    classes.update(tag.get("class"))  # Add all classes from the element

# Print the unique classes
print("Unique Classes in the HTML:")
for cls in sorted(classes):  # Sort for easier reading
    print(cls)
```
