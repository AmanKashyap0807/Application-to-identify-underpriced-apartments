```python
next_url = 'https://www.renthop.com/apartments-for-rent/new-york-ny?max_price=100000&page=2&sort=hopscore'
url_prefix = 'https://www.renthop.com/apartments-for-rent/new-york-ny?max_price=100000&page='
url_suffix = '&sort=hopscore'
```

```python
page_no =1
for i in range(1,3,1):
    target_url = url_prefix + str(i) + url_suffix
    print(target_url)
```

```python
def parse_data(listings_divs):
    listing_list = []

    for i in range(0, len(listings_divs), 2):
        current_listing = listings_divs[i]
        href = current_listing.select('a[id*=title]')[0]['href']
        addy = current_listing.select('a[id*=title]')[0].string.replace(' ','').replace('\n','')
        hood = current_listing.select('div[id*=hood]')[0].string.replace(' ','').replace('\n','')
        price = current_listing.select('div[id*=price]')[0].text.strip().replace('$','').replace(',','')
        beds = current_listing.select('div.font-size-10.d-inline-block.align-bottom[style="margin-left: 3px;"]')[0].text.strip().replace(" Bed", "")
        baths = listing_beds = listings_divs[0].select('div.font-size-10.d-inline-block.align-bottom[style="margin-left: 1px;"]')[0].text.strip().replace('Bath','').replace(' ','')
    
        listing_list.append([href, addy, hood, price, beds, baths])
        
    return listing_list
```

```python
def find_listings_divs(url):
    driver.get(url)
    page_s = driver.page_source
    soup = BeautifulSoup(page_s, 'html5lib')
    listings_divs = soup.select('div[class*=search-info]')
    
    return listings_divs
    
```

```python
final_listings_list = []
for i in range(1, 1000,1):
    target_url = url_prefix + str(i) + url_suffix
    listings_divs = find_listings_divs(target_url)
    listing_list = parse_data(listings_divs)
    final_listings_list.extend(listing_list)
    print(i)
```
